### numerical variabel

````python
numerical.head()
````

Ada banyak nilai nol pada numerical variabel, di mana hal itu biasanya menunjukkan adanya missing value dalam data. Akan tetapi, jika melakukan pendekatan berdasarkan konteks yang ada, di mana variabel is_canceled sebagai salah satu variabel yang memiliki banyak nilai nol, tentunya tidak masuk akal jika nilai nol mengindikasikan jumlah data yang hilang. Lebih logis jika nilai nol tersebut memang ada karena merupakan nilai yang sah.

Jika melihat variabel di sebelahnya pada 'data', yakni hotel dan lead_time. Pendekatan konteks serupa juga dapat dilakukan:
* Hotel: tidak mungkin mengandung nilai nol karena nilai-nilainya berupa kategori (teks) yang menunjukkan tipe hotel. Setiap entri pada variabel ini seharusnya berisi informasi hotel yang valid tanpa nilai nol, sehingga nol tidak menjadi masalah di sini.
* Lead_time: berdasarkan konteksnya, Variabel ini menunjukkan waktu (dalam hari) antara saat pemesanan dilakukan dan tanggal kedatangan. Jika ada nilai nol pada lead_time, kemungkinan besar artinya tamu melakukan pemesanan pada hari yang sama dengan kedatangan, dengan kata lain tidak ada selisih (hari) antara tanggal pemesanan dan kedatangan. Jadi, nilai nol di sini valid dan bukan indikasi missing value.

* Kemudian, variabel is_canceled sendiri secara kontekstualnya dapat terkait dengan pembatalan pesanan. Penambahan awalan is pada nama variabel juga biasanya ditujukan untuk pertanyaan yang mengharapkan 2 jawaban langsung seperti 'ya' atau 'tidak'. Dan umumnya penggunaan nilai biner pada bidang data, maka dapat diasumsikan bahwan ilai nol pada variabel ini juga bukan missing value, tetapi informasi sah yang menyatakan bahwa pemesanan berjalan sesuai rencana tanpa dibatalkan.  Nilai 0 di sini secara logis menunjukkan status "Tidak" (tidak dibatalkan), sedangkan 1 menunjukkan "Ya" (dibatalkan).

## missing value checked

````python
# persebaran nilai nan dalam dataset
missing = data.isnull().sum().sort_values(ascending=False)
percent = (missing/len(data)* 100).sort_values(ascending=False)
missing_value = pd.concat([missing, percent], axis=1, keys=['total', 'percent'])
missing_value
````

Seperti hasil yang didapatkan dari general information, terdapat 4 variabel yang memiliki missing value. Dengan variabel company yang 90% lebih datanya mengandung nilai yang hilang atau kosong. Selanjutnya, bagaimana sampai ada missing value dalam dataset?

Apakah karena memang sudah ada sejak awal atau terdapat kesalahan input dari pengguna? Error sistem? atau karena hal lain?

````python
nan_variabel = data[data.isnull().any(axis=1)]
nan_variabel.sample(50)
````

Pertama-tama, mari lihat persebaran data tidak nan-nya terlebih dahulu:

````python
not_nan_variable = data[data.notnull().all(axis=1)]
not_nan_variable.sample(50)
````

Dari kedua table di atas, variabel yang memiliki nilai nan dapat dibandingkan dengan yang tidak memiliki nilai nan, yang menghasilkan analisis berikut:

####1) **Why does the Company have missing value?**

* Dibandingkan dengan data yang memiliki nilai nan, data yang tidak memiliki nilai nan memiliki persentase yang lebih kecil. Hal ini jika dilihat sekilas tentu saja benar. Akan tetapi, banyaknya nilai nan itu sendirilah masalahnya.
* Pada variabel company, nilainya diisi dengan kode perusahaan yang melakukan pemesanan. Hal ini seperti "apakah (tamu) yang memesan adalah sebuah perusahaan?". Jika demikian, "jika pada variabel company diisi dengan yang bukan merupakan missing value, apakah berarti yang memesan reservasi adalah atas nama perusahaan?"
* Jika diamati, persebaran tersebut dapat diperkecil ruang lingkupnya dengan melihat hubungannya dengan variabel lain. Seperti, "siapa-siapa saja tamunya?". Hal tersebut dapat diketahui dengan melihat variabel children dan babies, dengan asumsi 'jika tamunya berasal dari perusahaan atau kalangan (orang) bisnis (contoh: pekerja kantoran)' maka secara logika tamu-tamu tersebut tidak akan membawa keluarga mereka karena harus fokus pada pekerjaannya, kecuali hal itu memungkinkan. Kemudian, pada variabel adults dan market_segment hal itu diperjelas dengan hanya sedikit orang dewasa yang merupakan tamu dari perusahaan di mana sebagian besar berasal dari segment 'coorporate'. Artinya, berpergian sendiri.

####2) **Why does the agent have missing value?**

````python
nan_variabel = data[data['agent'].isnull()]
nan_variabel.sample(50)
````

* Pertama, agent tidak setinggi company dalam persebaran missing value-nya, tetapi tetap menjadi perhatian utama. Seperti misalnya, ketika pada data yang tidak memiliki nan, sebagian besar data segment dan channel berpusat pada coorporate atau TA/TO, tetapi group, direct dan pilihan pemesanan sendiri tanpa melalui agent ketika berada di data yang memiliki missing value. Hal ini, tentu memberikan petunjuk bahwa sebagian besar tamu memiliki preferensi yang berbeda dalam melakukan reservasi kamar hotel.
* Di sisi lain, pemesanan di mana variabel agent-nya adalah missing value, dapat diartikan bahwa tamu/pelanggan tidak dalam jangkauan dari (kode) agent terkait,  yang mengindikasikan bahwa pelanggan memang lebih menyukai melakukan reservasi secara mandiri dan sistem agent hotel yang kurang dalam menjangkau pelanggan.

####3) **Why does the country have missing value?**

````python
nan_variabel = data[data['country'].isnull()]
nan_variabel.sample(50)
````

* Berdasarkan definisi country yang telah diambil sebelumnya, pada variabel ini adanya country dapat menunjukkan bahwa 'tamu tersebut dari negara b' atau 'negara asal seorang tamu melakukan reservasi', tetapi tidak adanya nilai pada variabel country tidak dapat diartikan bahwa 'tamu tersebut tidak berasal dari negara manapun' atau 'tamu tersebut tidak memiliki negara asal'. Bahkan jika terdapat [negara-negara yang tidak diakui oleh PBB](https://worldpopulationreview-com.translate.goog/country-rankings/countries-not-in-the-un?_x_tr_sl=en&_x_tr_tl=id&_x_tr_hl=id&_x_tr_pto=tc&_x_tr_hist=true), hal tersebut tidak dapat dijadikan patokan sebab konsep negara asal tetap berlaku dan tidak dapat dijadikan acuan pasti dalam analisis variabel ini.
* Karena itu, kemungkinan adanya tamu gelap dapat dihilangkan dan adanya tamu yang baik secara tidak atau dsengaja tidak mengisi negara asalnya juga dapat dipertimbangkan. Selain itu, jika tamu memesan reservasi melalui agent atau layanan bantuan pihak ketiga yang sejenisnya dan variabel country tetap tidak terisi, maka terdapat error dalam sistem pemesanan dan penginputan— karena terkait preferensi pelanggan/tamu yang mungkin saja 'ingin' menjadi annonim.

####4) **Why does the children have missing value?**

````python
nan_variabel = data[data['children'].isnull()]
nan_variabel
````

* Pada variabel children yang memiliki nilai nan, keempatnya merupakan pemesanan yang dibatalkan di selang waktu yang berdekatan. Selain itu, status reservasi nya pun di batalkan. Hal tersebut menunjukkan bahwa mungkin saja terjadi human error saat pelanggan melakukan reservasi. Kemungkinan itu tidak jauh dari definisi children 'adakah pelanggan yang merupakan bayi dalam daftar reservasi?'.
* Selain itu, nilai NaN pada variabel ini dapat disebabkan oleh pelanggan yang tidak mengisi informasi children karena reservasi dibatalkan lebih awal sebelum data diinput secara lengkap. Faktor lainnya adalah kurangnya validasi data dalam sistem selama proses pengisian formulir reservasi.
