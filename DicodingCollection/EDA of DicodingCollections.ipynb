Terdapat perbedaan waktu antara order_date dan delivery_date. Mengapa?

````python
delivery_time = orders["delivery_date"] - orders["order_date"]
delivery_time
````

Untuk menghitung waktu pengiriman, perlu dicari selisih antara tanggal produk di kirim dan tanggal pembuatan pemesanan. Pada kasus ini, dengan melakukan operasi pengurangan biasa selisih dalam bentuk hari sejatinya sudah didapatkan. Dengan menggunakan tipedata timedelta, yang merupakan representasi dari cara pandas terhadap durasi waktu, itu tidak sepenuhnya fleksibel terhadap analisis lain yang berkenaan dengan satuan konversi lain (seperti hari atau jam).

Karena itu, selisih hari yang telah di dapatkan, dalam hal ini, akan dikonversi menjadi satuan detik yang fleksibel terhadap satuan konversi lainnya.

Seperti contoh, jika mengambil index ke-0, di mana durasi pengiriman yang didapatkan ialah 25 hari. Jika dijabarkan:

````python
# satu menit itu 60 detik
# satu jam itu 60 menit
# dalam satu jam berarti ada 60 detik sebanyak 60.

# jika dilakukan operasi sederhana, maka:

a = 60  # detik dalam satu menit
b = 60  # menit dalam satu jam
c = a * b  # detik dalam satu jam
print(c)

# satu hari ada 24 jam, maka:
d = c * 24
print(d)
````
